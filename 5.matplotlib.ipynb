```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

- линейные графики

```python
# 1-мерный массив данных
x_arr = np.arange(-10,10)
x_arr
```

```python
y_arr = x_arr**2
```

```python
y_arr
```

```python
# визуализация
plt.figure(figsize=(10,8))
plt.plot(x_arr,y_arr)
plt.show()
```

```python
# совмещение нескольких графиков в одном окне
y_arr_1 = x_arr *2
y_arr_2 = x_arr**3
plt.plot(x_arr, y_arr_1)
plt.plot(x_arr, y_arr_2)
plt.show()
```

```python
# 1-мерный массив из 100 значений в диапазоне
x_arr_2 = np.linspace(-10,10,100)
x_arr_2
```

```python
# функция синуса
y_1 = np.sin(x_arr_2)
# функция гиперболического тангенса
y_2 = np.tanh(x_arr_2)
# функция сигмоиды
def sigmoid(x):
    return 1/(1+np.exp(-x))
y_3 = sigmoid(x_arr_2)
plt.plot(x_arr_2,y_1)
plt.plot(x_arr_2,y_2)
plt.plot(x_arr_2,y_3)
plt.show()
```

```python
# кастомизация линий (цвет)
y_1 = np.sin(x_arr_2)
y_2 = np.sin(x_arr_2 + 1)
y_3 = np.sin(x_arr_2 + 2)

# по названию цвета
plt.plot(x_arr_2, y_1, color="orange")

# по букве формата цвета из форматов RGB и CMYK
plt.plot(x_arr_2, y_2, color="b")

# по RGB 16коду или RGBA(прозрачность)
plt.plot(x_arr_2, y_3, color="#800000")

plt.show()
```

```python
# типы линий
# штриховая
plt.plot(x_arr_2, y_1, linestyle="dashed")

# пунктирная
plt.plot(x_arr_2, y_2, ls="dotted", lw=5)

# штрихпунктирная
plt.plot(x_arr_2, y_3, ls="dashdot", c="purple")

plt.show()
```

```python
# заголовок, легенда графика, подписи осей
plt.figure(figsize=(10,8))
y_1 = np.sin(x_arr_2)
y_2 = np.cos(x_arr_2)
plt.plot(x_arr_2,y_1)
plt.plot(x_arr_2,y_2)

#заголовок
plt.title("Заголовок")

#легенда графиков
plt.legend(["синус", "косинус"])

#подписи осей
plt.xlabel("ось x")
plt.ylabel("ось y")

plt.show()
```
