```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# загрузка данных excel (ctrl + пробел = выйдет подсказка после /) (с указанием страницы в excel)
df_0 = pd.read_excel("data/sred_potreb_cen_1991-2023.xlsx", sheet_name="1", skiprows=2, decimal=",")
df_0.head(10)
```

- очистка данных

```python
# удаление строк с пустыми значениями
df_1 = df_0.dropna()
df_1.head()
```

```python
df_1.shape
```

```python
df_1.info()
```

```python
# операция трансформирования таблицы (строки и столбы меняются местами)
df_1 = df_1.T
df_1.head()
```

```python
df_1.index
```

```python
# ключами делаем категории товаров и услуг (убираем их нумерацию)
df_1.columns = df_1.loc[' ПОКАЗАТЕЛИ ', :]
df_1.head()
```

```python
# удаление лишней строки
df_1 = df_1.drop(' ПОКАЗАТЕЛИ ')
df_1.head()
```

```python
df_1.columns
```

```python
# удаление лишних символов пробела перед и после названий ключей
df_1.columns = df_1.columns.str.strip(" ")
df_1.columns
```

```python
# попытка конвертации данных в тип Float
df_1 = df_1.astype("float")
```

```python
# замена '…' на 0
df_1 = df_1.replace('…', 0)
```

```python
df_1.info()
```

```python
# проверка столбца с ключом "..."
df_1["Сыры твердые, полутвердые и мягкие9)"].unique()
```

```python
# проверка столбца с ключом "..."
df_1["Водка 12), за л"].unique()
```

```python
# проверка столбца с ключом "..." (он был после 0. , я обновила) (пробелы изза пустых цифр после точки, которых должно быть 2 штуки)
df_1["Еврошифер21), за 10 м2"].unique()
```

```python
# замена '...' на 0
df_1 = df_1.replace('...', 0)
```

```python
# вторая попытка проверки
df_1 = df_1.astype("float")
```

```python
df_1.info()
```

```python
# сохранение в виде excel
df_1.to_excel("data/sred_potreb_cen_1991-2023.xlsx")
```

```python
# обрезка,срезка, удаление датафрейма
df_2 = df_1.loc[1998:, :]
df_2.head()
```

```python
# в сколько раз изменилась средняя цена за весь период
product = 'Масло сливочное 6)'
df_2.loc[2023, product] / df_2.loc[1998, product]
```
